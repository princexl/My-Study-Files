# 设置启动路径

## 方式一：修改配置文件

1. 命令行运行 `jupyter notebook --generate-config` 生成配置文件
2. 根据提示的路径找到文件夹 `.jupyter` 下的文件 `jupyter_notebook_config.py`
3. 找到 `#c.NotebookApp.notebook_dir = ''`，解除其注释并将其改为目标路径

## 方式二：修改快捷方式（推荐）

1. 找到 `jupyter notebook` 的快捷方式
2. 右键 —— 属性 —— 目标
3. 在 `%USERPROFILE%` 前添加目标路径，并留一个空格

# 给 Jupyter 添加扩展 

## 添加Nbextensions选项卡

[github上的教程](https://github.com/ipython-contrib/jupyter_contrib_nbextensions)

**<span class="burk">注意：Nbextensions选项卡和conda选项卡间似乎有冲突</span>**

1. 关掉 jupyter notebook（怎样都关不掉时，注销电脑），并在命令行中依次执行下列两条代码
```bash
pip install jupyter_contrib_nbextensions
jupyter contrib nbextension install --user
```
2. 重新启动`jupyter notebook`，菜单栏上多出`Nbextensions`选项卡

## 常用扩展
Nbextensions选项卡下的各种扩展

### AutoSaveTime
设置自动保存间隔

### ExecuteTime
在每一个code cell下显示执行消耗的时间及日期时间

### Datestamper
在工具栏上添加按钮，可以在任意cell中插入时间戳，如：2018-01-25 15:37:14 

### Codefolding
允许在code cell中折叠代码

### Codefolding和Codefolding in Editor
似乎功能重复，都是允许在code cell中折叠代码

### Freeze
可以为每个code cell设置模式
* normal：可以正常编辑执行
* read-only：仅可以执行，禁止编辑
* freeze：禁止编辑和执行

### Hide input
可以折叠选中的code cell的input

### Hide input all
同时折叠所有code cell的input

### Hinterland
code cell中输入代码时，原本只能通过按tab键的方式才可以联想自动补全，而安装这一扩展后，每一次按键都会进行提示自动补全。

### Printview
在新的浏览器窗口中打印预览

### Snippets Menu
在菜单栏上添加了Snippets菜单项，可以便捷地安装几个常用的数据分析包、获得它们的文档、并插入这些包的常用代码

### table_beautifier
美化markdown和html中出现的表格，并且能够对其按列进行排序，如：

|name|sex|age|
|--|--|--|
|a|b|c|
|d|e|f|
<span class="burk">但是在有些电脑上，指定对齐方式的功能似乎失效了，只能右对齐</span>

### 2to3 Converter
将code cell中python2的代码转化为python3的，<span class="mark">但是不知道为什么，图标并不是说明中的火箭，而是变成了锤子。</span>



### contrib_nbextensions_help_item
在jupyter菜单栏中Help一项下面添加了Jupyter-contrib，该链接指向扩展的目录及说明。

### Gist-it
Adds a button to publish the current notebook as a gist.

### Initialization cells
可以将code cell设置为initiallization cell，并使其在该notebook被打开时自动执行

<span class="burk">在thinkpad上莫名其妙地无效。</span>

### Python Markdown
允许在markdown cell中通过双花括号括住的方式嵌入python的对象

显示全部code cell中代码的行号

### Help panel

# 常见 error

## kernel error (kernel not found)
【原因】`python` 编译器的路径不正确

【解决方案】
1. 命令行内运行命令 `jupyter kernelspec list` 查看安装的内核和位置
2. 进入安装内核目录打开 `kernel.json` 文件，查看 `Python` 编译器路径
3. 如果路径错误，要重新安装内核，命令行内运行命令 `python -m ipykernel install --user`

【注意】
1. 安装内核时，`python` 编译器的名字要写正确（比如我将 `python2` 的编译器 `python.exe` 重命名为了 `python27.exe`，则命令行内运行命令应为 `python27 -m ipykernel install --user`）
2. 当给 `jupyter` 安装了一些扩展程序（比如RISE）后，启动时加载kernel的路径可能会发生改变，需要将 `kernel.json` 文件复制到对应路径。

````python
!jupyter kernelspec list
````

## jupyter notebook中conda选项卡error
【原因】`conda` 环境中的配置文件和用户目录内的配置文件冲突。

【解决方案】删除用户目录中 `.jupyter` 文件夹内的文件 `jupyter_notebook_config.json`

# 改变jupyter的主题

## 法一
在第一个code cell中输入以下代码并运行，所有的cell都会显示黑色背景。

```javascript
%%javascript

window.load_remote_theme = true
var theme_js = "https://odhk.github.io/hyrule_theme/custom.js";

window.load_local_theme = function(){
    var hostname = document.location.hostname
    return ((hostname == "localhost" || hostname == '127.0.0.1') && !load_remote_theme)
}

var url = load_local_theme() ? document.location.origin + "/files/theme/custom.js" : theme_js
$.getScript(url)
```

## 法二 
1. 安装 `jupyter-themer` —— `pip install jupyter-themer`
2. 命令行启动 `jupyter` —— `jupyter-themer -c COLOR -f CODE_FONE`，（在命令行输入 `jupyter-themer --show color` 查看可选颜色），更多指令请查询 [transcranial/jupyter-themer](https://github.com/transcranial/jupyter-themer)

## 法三
upyterthemes模块，详见 [改变jupyter notebook的主题.ipynb](改变jupyter notebook的主题.ipynb)
