- **将DataFrame中[lgt,lat]形式的坐标拆分成lgt,lat各单独一列**

```python
import pandas as pd
from pandas import DataFrame

# 首先有一个含有[lgt,lat]形式坐标的df
baoshan = {"hid":[0,1,2,3,4],
           "bd_center":["[11,111]","[22,222]","[33,333]","[44,444]","[55,555]"]}
baoshan = DataFrame(baoshan)
print(baoshan)
```

```python
# 注意这里经纬度为string
type(baoshan["bd_center"][0])
```

```python
# 目标是将经度纬度分别提取出来，放入对应的一列中
baoshan['bd_center'] = baoshan['bd_center'].apply(eval)#  先将经纬度转化为列表
type(baoshan["bd_center"][0])
```

```python
# 对于"bd_center"的每一个lgt，都放在新建的"lgt"列里；lat同理
baoshan['lgt'] = baoshan['bd_center'].apply(lambda x: x[0])
baoshan['lat'] = baoshan['bd_center'].apply(lambda x: x[1])
print(baoshan)
```

```python
import pandas as pd
from functools import reduce

# 读取数据
loc_14000 = pd.read_csv(r'F:\Cardinal\Project\Yonghui\stage2\result\result_0204\suggested_locations_14000_0204.csv')
# 提取hid列
hid_14000 = pd.DataFrame(loc_14000['hid'])

# 读取方格邻居数据，每个方格有8个邻居，九个hid在一个list里
neighbor = pd.read_csv('F:\Cardinal\Project\Yonghui\stage2\stability\grids_with_neighbors.csv')
# 对每个hid匹配其邻居
neighbor_14000 = pd.merge(hid_14000, neighbor, left_on='hid', right_on='hid')

# 将每种结果的hid list整合为集合
neighbor_14000['neighbors'] = neighbor_14000['neighbors'].apply(eval) # 将字符串变为set
list1 = neighbor_14000['neighbors'].tolist() # 将一列中所有的set合并为一个大list，共397项

# 用reduce函数对list中的每一项进行累加取并集，得到所有hid的大集合
set1 = reduce((lambda x, y: x | y), list1)
print(len(set1))
```
